```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('dados/resultado-00000-of-00001.csv', delimiter=';')
df
```

```python
df_group_by = df.groupby(['UF', 'YEAR'])[['RAINFALL', 'DENGUE_CASES']].mean().round(2).reset_index()
df_group_by.head()
```

```python
df_CE = df_group_by.loc[lambda x: x['UF'] == 'CE']
df_ES = df_group_by.loc[lambda x: x['UF'] == 'ES']
df_MG = df_group_by.loc[lambda x: x['UF'] == 'MG']
df_PR = df_group_by.loc[lambda x: x['UF'] == 'PR']
df_RJ = df_group_by.loc[lambda x: x['UF'] == 'RJ']
df_RS = df_group_by.loc[lambda x: x['UF'] == 'RS']
df_SP = df_group_by.loc[lambda x: x['UF'] == 'SP']
```

```python
ax = sns.pointplot(df_CE, x='YEAR', y='DENGUE_CASES', markers='o')
ax.set_title('Media de casos de dengue x Ano no estado do Ceara')
ax.set_xlabel('Ano')
ax.set_ylabel('Casos de dengue')
ax.figure.set_size_inches(12,6)
ax
```
